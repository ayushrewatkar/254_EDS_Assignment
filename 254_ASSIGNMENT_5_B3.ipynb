```python
import matplotlib.pyplot as plt
import pandas as pd
df=pd.read_csv("/content/tips.csv")
print(df)
c1=["red","blue","green","orange"]
c2=["red","green"]
t1=df.groupby("day").count()
t2=df.groupby("time").count()
t3=df.groupby("size").count()
t4=df.groupby("sex").count()
t5=df.groupby("smoker").count()
```

```python
c2=["red","blue"]
t4["tip"].plot(kind="bar",color=c2)
plt.title("Sex vs Tip")
```

```python
t5["time"].plot(kind="bar",color=["red","green"])
plt.title("Smoker vs Time")
```

```python
t3["sex"].plot(kind="pie",autopct="%1.f%%")
plt.title("Size vs Sex")
```

```python
t1["sex"].plot(kind="bar", color=c1)
plt.title("Day vs Sex")
```

```python
t1["sex"].plot(kind="pie",autopct="%1.f%%")
plt.title("Day vs Sex")
```

```python
c3=["blue","red"]
t1["smoker"].plot(kind="bar",color=c3)
plt.title("Day vs Smoker")
```
