```python
import pandas as pd
df=pd.read_csv("/content/grains.csv")
#print("Data",df)
print("######################################################################################################")
#1 which type of community found in Chittor district?
community=df[df['district'] == 'Chittor']['commodity'].unique()
print("1] the community founded in chittor district is ",community)
#2 how many varieties are local?
local= len(df[df['variety'] == 'Local'])
print("2] there are",local,"varieties are local")
#3 sum of modal price of Carrot and onion?
total= df[df['commodity'].isin(['Carrot', 'Onion'])]['modal_price'].sum()
print("3] the sum of modal price of carrot and onion is ",total)
#4 average modal price of Cauliflower?
avg=df[df['commodity'] == 'Cauliflower']['modal_price'].mean()
print("4] the average of modal price of Cauliflower is",avg)
#5 print the data have same district and market?
print("5]",df.loc[df['district'] == df['market']])
#6 Find the no of grains in Andaman and Nicobar which have minimum price greater than 6000?
filtered_data = df[(df['state'] == 'Andaman and Nicobar') & (df['min_price'] > 6000)]
num_grains = len(filtered_data)
print("6] Number of grains with minimum price greater than 6000 in Andaman and Nicobar:", num_grains)
#7 Which grain get more price in this data set?
highest_price_row = df[df['max_price'] == df['max_price'].max()]
Max_Price = df["max_price"].max()
grain_with_highest_price = highest_price_row.iloc[0,3]
print("7] Grain with the highest price:", grain_with_highest_price,'=',Max_Price)
#8 find the no. of types of grains in this dataset?
filtered_data = df.groupby("commodity").count()
num_grains = len(filtered_data)
print("8] Number of types of grains in this dataset is:", num_grains)
#9 What is the average modal price across all commodities in the dataset?
average_modal_price = df['modal_price'].mean()
print("9] Average modal price across all commodities:", average_modal_price)
#10 How many different districts are represented in the dataset?
num_district = df["district"].nunique()
print("10] Number of different districts are represented in the dataset is:",num_district)
#11 What is the most common variety of rice in the dataset?
rice_data = df[df["commodity"]=="Rice"]
most_common_rice_variety = rice_data['variety'].mode().values[0]
print("11] The most common variety of rice in the dataset is:",most_common_rice_variety)
#12 How many commodities have a modal price greater than 10,000?
filtered_data = df[df['modal_price'] > 10000]
num_commodities = len(filtered_data)
print("12] Number of commodities with a modal price greater than 10,000:", num_commodities)
#13 Which district has the lowest minimum price for potatoes?
potato_data = df[df['commodity'] == 'Potato']
district_lowest_price = potato_data[potato_data['min_price'] == potato_data['min_price'].min()]['district'].values[0]
print("13] The district has the lowest minimum price for potatoes is:", district_lowest_price)
#14 What is the average maximum price for bananas across all districts
Bananas_data = df[df['commodity'] == 'Banana']
average_max_price = df['max_price'].mean()
print("14] Average maximum price for bananas across all districts is:", average_max_price)
#15 Which state has the highest number of commodities listed in the dataset?
state_commodity_counts = df['state'].value_counts()
state_with_highest_commodities = state_commodity_counts.idxmax()
print("15] State with the highest number of commodities listed:", state_with_highest_commodities)
#16 How many unique districts are there in the dataset?
print("16] Number of unique districts:", df['district'].nunique())
#17 What is the average modal price for the commodity 'Rice' across all districts?
rice = df[df['commodity'] == 'Rice']
ave = rice['modal_price'].mean()
print("17] Average modal price for Rice:", ave)
#18 What is the range of prices for the commodity 'Jute' in each state?
jute = df[df['commodity'] == 'Jute']
price = jute.groupby('state')['max_price'].max() - jute.groupby('state')['min_price'].min()
print("18] Price range for Jute in each state:",price)
#19 On which date did the commodity 'Pumpkin' have the highest modal price?
pumpkin=df[df['commodity'] == 'Pumpkin']
high = pumpkin.loc[pumpkin['modal_price'].idxmax(), 'arrival_date']
print("19] Date with the highest modal price for Black pepper:", high)
#20 What is the total number of arrivals of all commodities in each state?
arrivals = df.groupby('state')['arrival_date'].count()
print("20] Total number of arrivals of all commodities in each state:",arrivals)
print("######################################################################################################")
```
